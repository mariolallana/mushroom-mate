```python
import sqlite3
import mysql.connector

def migrate_sqlite_to_mysql(sqlite_db_path, mysql_params):
    # Conectar con la base de datos SQLite
    sqlite_conn = sqlite3.connect(sqlite_db_path)
    sqlite_cursor = sqlite_conn.cursor()

    # Conectar con la base de datos MySQL
    mysql_conn = mysql.connector.connect(**mysql_params)
    mysql_cursor = mysql_conn.cursor()

    # Obtener las tablas de la base de datos SQLite
    sqlite_cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = sqlite_cursor.fetchall()

    for table in tables:
        table_name = table[0]
        # Obtener los datos de la tabla
        sqlite_cursor.execute(f"SELECT * FROM {table_name};")
        rows = sqlite_cursor.fetchall()

        # Crear la tabla en MySQL
        create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} ("
        sqlite_cursor.execute(f"PRAGMA table_info({table_name})")
        columns_info = sqlite_cursor.fetchall()
        for i, col_info in enumerate(columns_info):
            column_name = col_info[1]
            column_type = col_info[2]
            create_table_query += f"{column_name} {column_type}"
            if i < len(columns_info) - 1:
                create_table_query += ", "
        create_table_query += ");"
        print("Create Table Query:", create_table_query)
        mysql_cursor.execute(create_table_query)

        # Insertar datos en la tabla MySQL
        for row in rows:
            placeholders = ', '.join(['%s'] * len(row))
            insert_query = f"INSERT INTO {table_name} VALUES ({placeholders});"
            print("Insert Query:", insert_query, "Values:", row)
            mysql_cursor.execute(insert_query, row)

    # Confirmar los cambios en MySQL
    mysql_conn.commit()

    # Cerrar las conexiones
    sqlite_cursor.close()
    sqlite_conn.close()
    mysql_cursor.close()
    mysql_conn.close()

# Parmetros de conexin a MySQL
mysql_params = {
    "host": "localhost",
    "user": "mario",
    "password": "chetler2",
    "database": "forest"
}

# Ruta de la base de datos SQLite
sqlite_db_path = "C:/Users/mario/mushroom-mate/backend/forest_data.db"

# Ejecutar la migracin
migrate_sqlite_to_mysql(sqlite_db_path, mysql_params)
```
